```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense
```

```python
cancer = datasets.load_breast_cancer()
```

```python
X_train_roc, X_test_roc, y_train_roc, y_test_roc = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state = 50)
```

```python
bpnn = keras.Sequential([
    Dense(10, input_shape = (30,), activation = 'relu'),
    Dense(3, activation = 'sigmoid')
])

bpnn.compile(optimizer = 'Adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
bpnn.fit(X_train_roc, y_train_roc, epochs = 1000)
```

```python
bpnn.evaluate(X_test_roc, y_test_roc)
```
